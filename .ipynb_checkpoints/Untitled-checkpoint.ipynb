## Find the two numbers that sum to 2020 and calculate their product

```python
nums = pd.read_csv('nums.csv', header=None)
nums = list(nums.iloc[:,0].values)
nums
```

```python
def find_sum(lst):
    for i in lst:
        for j in lst:
            if i + j == 2020:
                print(i,j)
                return i*j
    return 'None found'
```

```python
find_sum(nums)
```

```python
def find_sum_three(lst):
    for i in lst:
        for j in lst:
            for k in lst:
                if i + j + k == 2020:
                    print(i,j,k)
                    return i*j*k
    return 'None found'
```

```python
find_sum_three(nums)
```

## Find valid passwords
A valid password contains the specified number of letters in the beginning of the string. The password chosen is after the colon (:)
Modified: The min and max are actually positions within the password. Need to have given letter in EITHER the min OR the max, NOT both (exclusive or).

```python
passwords = pd.read_csv('advent2.csv', header=None)
passwords = list(passwords.iloc[:,0].values)
passwords
```

```python
p = passwords[0]
tup = p.split(' ')
req_min = int(tup[0].split('-')[0])
req_max = int(tup[0].split('-')[1])
req_letter = tup[1][0]
password = tup[2]
num_letter = password.count(req_letter)
req_letter, req_min, req_max, password
if (req_min > num_letter) or (req_max < num_letter):
    print(0)
else:
    print(1)
```

```python
def split_item(item):
    tup = item.split(' ')
    req_min = int(tup[0].split('-')[0])
    req_max = int(tup[0].split('-')[1])
    req_letter = tup[1][0]
    password = tup[2]
    
    return req_letter, req_min, req_max, password

def find_valid_passwords(lst):
    n_valid = 0
    n_invalid = 0
    for p in passwords:
        # Get the letter, min, max, and password
        req_letter, req_min, req_max, password = split_item(p)
        num_letter = password.count(req_letter)
        
        # CHeck if valid
        if (req_min > num_letter) or (req_max < num_letter):
            n_invalid += 1
        else:
            n_valid += 1
    return n_valid, n_invalid
        
def find_valid_passwords_modified(lst):
    n_valid = 0
    n_invalid = 0
    for p in passwords:
        # Get the letter, min, max, and password
        req_letter, req_min, req_max, password = split_item(p)
        
        # Check if valid. The ^ is exclusive or
        if (password[req_min-1] == req_letter) ^ (password[req_max-1] == req_letter):
            n_valid += 1
        else:
            n_invalid += 1
    return n_valid, n_invalid
```

```python
find_valid_passwords(passwords)
```

```python
find_valid_passwords_modified(passwords)
```

## Day 3: Find the number of trees encountered
Given a grid of trees on a ski slope, count the # of trees encountered if moving 3 to the right and 1 down (consecutively)
Part 2: Multiply all tress encountered for the following:

Right 1, down 1.
Right 3, down 1. (This is the slope you already checked.)
Right 5, down 1.
Right 7, down 1.
Right 1, down 2.

```python
ski = np.loadtxt('advent3.txt', dtype=str, delimiter=None)
ski = list(ski)
```

```python
f = open('advent3.txt', 'r')
ski = f.readlines()
cleaned = [s.strip('\n') for s in ski]
```

```python
def find_trees(grid, right, down):
    n_trees = 0
    curr_row = 0
    curr_col = 0
    repeat_size = len(grid[0])
    bottom_row = len(grid)

    while curr_row < bottom_row:
        # Update # trees
        n_trees += grid[curr_row][curr_col] == '#'
        
        # Update row and col
        curr_col = (curr_col + right) % repeat_size
        curr_row += down
    
    return n_trees
```

```python
[find_trees(cleaned, r, d) for r, d in zip([1,3,5,7,1], [1,1,1,1,2])]
```

## Day 4: Find valid passports
Part 1: Find # passports that are valid (have all required entries)

Part 2: Find # passports that have valid entries and valid values for those entries

byr (Birth Year) - four digits; at least 1920 and at most 2002.
iyr (Issue Year) - four digits; at least 2010 and at most 2020.
eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
hgt (Height) - a number followed by either cm or in:
If cm, the number must be at least 150 and at most 193.
If in, the number must be at least 59 and at most 76.
hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
pid (Passport ID) - a nine-digit number, including leading zeroes.
