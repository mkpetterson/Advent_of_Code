```python
def find_sum(lst):
    for i in lst:
        for j in lst:
            if i + j == 2020:
                print(i,j)
                return i*j
    return 'None found'
```

A valid password contains the specified number of letters in the beginning of the string. The password chosen is after the colon (:)

```python
def split_item(item):
    tup = item.split(' ')
    req_min = int(tup[0].split('-')[0])
    req_max = int(tup[0].split('-')[1])
    req_letter = tup[1][0]
    password = tup[2]
    
    return req_letter, req_min, req_max, password

def find_valid_passwords(lst):
    n_valid = 0
    n_invalid = 0
    for p in passwords:
        # Get the letter, min, max, and password
        req_letter, req_min, req_max, password = split_item(p)
        num_letter = password.count(req_letter)
        
        # CHeck if valid
        if (req_min > num_letter) or (req_max < num_letter):
            n_invalid += 1
        else:
            n_valid += 1
    return n_valid, n_invalid
        
def find_valid_passwords_modified(lst):
    n_valid = 0
    n_invalid = 0
    for p in passwords:
        # Get the letter, min, max, and password
        req_letter, req_min, req_max, password = split_item(p)
        
        # Check if valid. The ^ is exclusive or
        if (password[req_min-1] == req_letter) ^ (password[req_max-1] == req_letter):
            n_valid += 1
        else:
            n_invalid += 1
    return n_valid, n_invalid
```

## Day 3: Find the number of trees encountered
Given a grid of trees on a ski slope, count the # of trees encountered if moving 3 to the right and 1 down (consecutively)
Part 2: Multiply all tress encountered for the following:

Right 1, down 1.
Right 3, down 1. (This is the slope you already checked.)
Right 5, down 1.
Right 7, down 1.
Right 1, down 2.

```python
f = open('advent3.txt', 'r')
ski = f.readlines()
cleaned = [s.strip('\n') for s in ski]
```

```python
def find_trees(grid, right, down):
    n_trees = 0
    curr_row = 0
    curr_col = 0
    repeat_size = len(grid[0])
    bottom_row = len(grid)

    while curr_row < bottom_row:
        # Update # trees
        n_trees += grid[curr_row][curr_col] == '#'
        
        # Update row and col
        curr_col = (curr_col + right) % repeat_size
        curr_row += down
    
    return n_trees
```

## Day 8: Find accumulator value after first set of instructions 

```python
import re
f = open('advent8.txt')
lines = f.readlines()
lines = [f.strip('\n') for f in lines]
lines
```

```python
def execute(n, data, instruction_count, accumulator):
    # Define base case
    instruction_count[n] += 1
    if 2 in instruction_count.values():
        return accumulator
    
    # Get number
    item = data[n]
    instruction_sign = re.findall('[\+\-]', item)[0]
    instruction_number = int(re.findall('\d+', item)[0])
        
    # Execute 
    if re.match('acc', item):
        accumulator += instruction_number
    elif re.match('jmp', item):
        new_idx = n + instruction_number
        execute(new_idx, data, instruction_count, accumulator)
    else:
        execute(n+1, data, instruction_count, accumulator)
    
    return 'no loop'

def main(data):
    # Get number of entries and create structure to keep track of what has been run
    length = len(lines)
    instruction_count = {k:0 for k in range(length)}
    accumulator = 0
    
    accumulator = execute(0, data, instruction_count, accumulator)

    return accumulator
```

## Day 5
Find the highest ID on the bording passes


```python
f = open('advent5.txt', 'r')
data_raw = f.readlines()
boarding_passes = [line.strip('\n') for line in data_raw]
boarding_passes
```

```python
def convert_string(substring, stack):
    #print('starting, stack length is: {}'.format(len(stack)))
    # Base case
    if len(stack) == 1:
        return stack[0]

    # Find middle point of stack
    midpoint = int(len(stack)/2)

    # Get first char and remove from substring
    substring = list(substring)
    char = substring.pop(0)
    #print(stack, midpoint, char, substring)

    # Get recursive with first or second half of list
    if char == 'F':
        return convert_string(substring, stack[:midpoint])
    else:
        return convert_string(substring, stack[midpoint:])


def calc_id(row, col):
    return row*8 + col


def main(passes):
    seat_ids = []
    # Get individual seat ids for each ticket
    for ticket in passes:
        # Parse out row and col identifiers and get row, col numbers
        row_id, col_id = ticket[:7], ticket[7:]
        row = convert_string(list(row_id), list(range(0,128)))
        col = convert_string(list(col_id), list(range(0,8)))
        seat_id = calc_id(row, col)
        seat_ids.append(seat_id)
        
        #print(row, col, seat_id)
        
    return max(seat_ids)
```

```python
main(boarding_passes)
```
